```python
from tensorflow.keras.datasets.mnist import load_data
(x_train, y_train), (x_test, y_test) = load_data()
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt
print(y_train[0])
# cmap = "_r"相反
plt.imshow(x_train[0], cmap="gray_r")
```

```python
import pandas as pd
pd.DataFrame(x_train[0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

layers = [
    # 784 * 128 (線) + 128 (bias) =100480
    Dense(256, activation="relu", input_dim=784),
    # 128 * 10 (線) + 10 = 1290
    Dense(10, activation="softmax")             
]

model = Sequential(layers)
model.summary()
```

```python
# mse 的梯度變化小, corss entropy 梯度變化大
# 二元()
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
model.compile(
    loss=CategoricalCrossentropy(),
    optimizer = Adam(),
    metrics=["accuracy"]
)
```

```python
# 資料預處理
# 答案: [0, 5, 2 1,,..]
from tensorflow.keras.utils import to_categorical
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
print(y_train[0])
print(y_train_cat[0])
```

```python
# 圖片 : [255, 0, 255, 128]
# 28 * 28  >> 784
x_train_norm = x_train.reshape(60000, 784) / 255.0
x_test_norm = x_test.reshape(10000, 784) / 255.0
# print(x_train_norm[0])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
# valodation_split : 切出驗證的資料來做驗證 (0.1 : 540000筆資料訓練)
# epochs : 看幾輪  (10輪 : 54000 * 10)
# batch_size : 看了多少筆再做一次梯度調整(看圖片大小  ,小圖:200, 大圖:20)
# 200 batch, 10 epoch : 540000 /200 >> 2700次
# verbose : 0(無聲), 1(更多訊息), 2(沒有進度條)
callbacks = [
            EarlyStopping(patience=5, restore_best_weights=True),
            ModelCheckpoint("mlp.h5", save_best_only=True)
]

model.fit(x_train_norm,
          y_train_cat,
          validation_split = 0.1,
          epochs = 100,
          batch_size = 200,
          verbose = 2,
          callbacks=callbacks)
```

```python
model.evaluate(x_test_norm, y_test_cat,verbose=2)
```

```python
from sklearn.metrics import confusion_matrix
# predict proba -> predict
# predict -> predict classes
# 只要在 Tensorflow 就帶  one-hot y_test cat
# 不在,就帶原本  y_test
pre = model.predict_classes(x_test_norm)
mat = confusion_matrix(y_test, pre)
pd.DataFrame(mat, 
             columns=["{}(預測)".format(i) for i in range(10)],
             index=["{}(正確)".format(i) for i in range(10)])
```

```python
import numpy as np
idx = np.nonzero(pre != y_test)[0]
# np.nonzero([[0, 2, 0], [0, 0, 3]])
false_pre = pre[idx]
false_label = y_test[idx]
false_img = x_test[idx]

plt.figure(figsize=(14, 42))
width = 10
height = len(idx) // width + 1
for i in range(len(false_img)):
    plt.subplot(height, width, i+1)
    title = "[P]:{}\n[O]:{}".format(false_pre[i], false_label[i])
    plt.title(title)
    plt.axis("off")
    plt.imshow(false_img[i], cmap="gray_r")
```
